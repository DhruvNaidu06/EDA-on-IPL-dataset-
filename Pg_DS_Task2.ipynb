```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
#histogram for player_of_match

plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]))
plt.show()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
# Making a histogram for Distribution of Runs batting first

plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
# Making a bar chart for top 3 teams batting first

plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","red"])
plt.show()
```

```python
#Pie chart for Batting First and winning

plt.figure(figsize=(8,8))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
# Making a histogram for team batting second

plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=25)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
# Making a bar chart for top 3 teams batting second

plt.figure(figsize=(6,6))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#Pie chart for Batting Second with winning percentage

plt.figure(figsize=(8,8))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
